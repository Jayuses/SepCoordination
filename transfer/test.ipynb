```python
import rdkit
import numpy as np
import pandas as pd
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
```

```python
smiles = pd.read_excel('../dataset/Data/ourdata/Cu_graph.xlsx',0,usecols='AQ').to_numpy().squeeze().tolist()
```

```python
m = smiles[0]
m = Chem.MolFromSmiles('CC1=CC=C(C=C1)NC2=C3C(=C(C=C2)NC4=CC=C(C=C4)C)C(=O)C5=CC=CC=C5C3=O')
m3d = Chem.AddHs(m)
AllChem.EmbedMolecule(m3d, AllChem.ETKDGv3())
Draw.MolToImage(m3d, size=(250,250))
```

```python
conformer = m3d.GetConformer()
coor = conformer.GetPositions()
```
